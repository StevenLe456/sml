# Statistical Machine Learning Overview

## Brief Overview
* Statistical machine learning algorithms are inductive algorithms (meaning they generate plausible inferences from an input pattern using a knowledge base of beliefs)
    * Statistical machine learning is based on the framework of probabilistic inductive logic.
    * Training data is generated from the environment by sampling from an environmental probability distribution
    * The process that generates the training data is called the data generating process
    * A statistical learning machine's knowledge base is a set of probability distribution called the probability model
* The goal of statistical machine learning is not to memorize the data
    * The goal is to learn the data generating process by finding the probability distribution in the learning machine's probabilty model which best approximates the environmental distribution
    * The learning machine then uses this best approximating distribution to make inferences on other data
![Diagram of SML framework](../img/sml_1.png)

```python
# Snippet of code to simulate the statistical machine learning framework
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Environmental distribution is the standard normal distribution

# Data generating process samples data from the environmental distribution (standard normal distribution)
def data_generating_process(num_samples):
    return np.random.normal(0, 1, size=num_samples)

x = np.arange(-10, 10, 1)

# The probability model is a set of probability distributions
probability_model = [norm.pdf(x, 0, 1), norm.pdf(x, -5, 2), norm.pdf(x, 5, 0.5)]

# The goal of the learning machine is to search for the best approximate distribution for the environmental distribution
# based on the data generating process within the probability model

# Having the learning machine search for an approximate distribution is way out of the scope of this notebook right now,
# so I will plot the training sample and the probability model to show the data generating process and probability model visually
training_data = data_generating_process(10000)
plt.hist(training_data, bins=100)
bin_width = (training_data.max() - training_data.min()) / 100
hist_area = len(training_data) * bin_width
plt.plot(x, probability_model[0] * hist_area)
plt.plot(x, probability_model[1] * hist_area)
plt.plot(x, probability_model[2] * hist_area)
plt.show()
```
